```python
def get_average_cost_subitem(item: str) -> float:
    return df_mat.loc[df_mat.Item == item, 'Average energy cost'].values[0]
```

```python
n = 1
cost = 0
subitem_name = df_mat.loc[df_mat.Item == item, 'Item ' + str(n)].values[0]
subitem_qty = df_mat.loc[df_mat.Item == item, 'Quantity ' + str(n)].values[0]

    
while (n < 3) and (type(subitem_name) == str):
    cost += get_average_cost_subitem(subitem_name) * subitem_qty
    print(n, subitem_name, cost)
    
    n += 1
    subitem_name = df_mat.loc[df_mat.Item == item, 'Item ' + str(n)].values[0]
    subitem_qty = df_mat.loc[df_mat.Item == item, 'Quantity ' + str(n)].values[0]

    
```

```python
subitem_name
```

```python
def get_average_cost_craftable(item, df_mat):
    n = 1
    cost = 0
    subitem_name = df_mat.loc[df_mat.Item == item, 'Item ' + str(n)].values[0]
    subitem_qty = df_mat.loc[df_mat.Item == item, 'Quantity ' + str(n)].values[0]

    while (n < 3) and (type(subitem_name) == str):
        cost += get_average_cost_subitem(subitem_name) * subitem_qty
        # print(n, subitem_name, cost)
        n += 1
        subitem_name = df_mat.loc[df_mat.Item == item, 'Item ' + str(n)].values[0]
        subitem_qty = df_mat.loc[df_mat.Item == item, 'Quantity ' + str(n)].values[0]

    return cost
```

```python
df_mat.loc[df_mat['Craftable'] == True, 'Average energy cost'] = df_mat.loc[df_mat['Craftable'] == True, 'Item'].apply(get_average_cost_craftable, df_mat=df_mat)
```

# Try to do an item

```python
item = 'Adamantium Blades'
# item = 'Adamantium Lump'
```

```python
row = df_mat.loc[df_mat.Item == item]
row
```

```python
from bigtree import Node, print_tree
from math import isnan
```

```python
df_mat['Craftable'].loc[df_mat.Item == row.iloc[0, 4 + 0*2]].iloc[0]
```

```python
row.iloc[0, 3] # Craftable
row.iloc[0, 4] # Item 1
row.iloc[0, 5] # Quantity 1
```

```python
root = Node(row.Item.iloc[0], 
            craftable=row['Craftable'].iloc[0],
            quantity=1)
for n in [0, 1, 2]:
    if not isnan(row.iloc[0, 5 + n*2]): # sub-item name
        child = Node(row.iloc[0, 4 + n*2],
                     craftable=df_mat['Craftable'].loc[df_mat.Item == row.iloc[0, 4 + n*2]].iloc[0],
                     quantity=row.iloc[0, 5 + n*2], 
                     parent=root)
        child.quantity = child.quantity * child.parent.quantity
        if child.craftable == False:
            child_data = df_raw.loc[df_raw['Item'] == child.name, ['Campaign', 'Variation', 'Level', 'Average energy cost']].sort_values(by=['Average energy cost'])
            # child.location = df_raw
            child.cost = child_data.iloc[0]['Average energy cost']
        
print_tree(root, attr_list=["craftable", "quantity", "cost"])
```
