```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

# Load and check the data

```python
PATH_MAT = './data/materials_cleaned.csv'
df = pd.read_csv(PATH_MAT, index_col=0)
df
```

```python
df.loc[df['Average energy cost'].isna()]
```

```python
fig, axs = plt.subplots(2, 2)

for i, (ax, rarity) in enumerate(zip(axs.flatten(), ["Uncommon", "Rare", "Epic", "Legendary"])):
    sns.histplot(data=df[df[["Craftable", "Rarity"]].eq([True, rarity]).all(1)], 
                x="Average energy cost",
                ax=ax)
    ax.set_title(rarity)
plt.tight_layout()
plt.show()
```

```python
# fig, axs = plt.subplots(4, 1, sharex=True, sharey=True, dpi=300, figsize=(4,10))
fig, axs = plt.subplots(3, 1, sharex=True, sharey=True, dpi=150, figsize=(4,5))

for i, (ax, rarity) in enumerate(zip(axs.flatten(), ["Rare", "Epic", "Legendary"])):
    data = df[df[["Craftable", "Rarity"]].eq([True, rarity]).all(1)]
    sns.histplot(data=data, 
                x="Average energy cost",
                ax=ax)
    ax.set_title(rarity)
    ax.axvline(data["Average energy cost"].mean())
plt.tight_layout()
plt.show()
```
